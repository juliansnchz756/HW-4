```python
import matplotlib.pyplot as plt
import numpy as np

marks = {'Andy':88, 'Amy':66, 'James': 90, 'Jules': 55, 'Arthur': 77}

name = list(marks.keys())

grade = list(marks.values())

plt.bar(name,grade)
plt.show()

plt.pie(grade, labels = name)
plt.legend(bbox_to_anchor=(1.05,1.0),loc='upper left')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

numbers = np.array([[1,2,3],[5,6,7]])

plt.plot(numbers)
plt.show()

plt.bar(numbers[0], numbers[1])
plt.show()

plt.pie(numbers[0], labels = numbers[1])
plt.legend(bbox_to_anchor=(1.05,1.0),loc='upper left')
plt.show()

plt.scatter(numbers[1], numbers[0])
plt.show()

plt.hist(numbers)
plt.show()
```
